##### Min-Max Scaling: MinMax scaling transforms the data into a specific range, typically [0, 1] or [-1, 1], ensuring that all features have the same scale. This is particularly useful when you want to preserve the relationships between the original data points and ensure that the scaled values fall within a predefined range. It's beneficial when you want to compare or visualize different features on a common scale. For stock price prediction, MinMax scaling can be helpful if you have features with vastly different value ranges, as it prevents one feature from dominating the others in the modeling process

```python
## normalization for A1 stock using z score normalization
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,MinMaxScaler
# Define the columns to normalize
normalize_col=["Opening Price","High","Low","Closing Price","Volume"]
label=["Closing Price"]
# Extract the selected columns for the features (X) and the target (y)
stock_X=dfA1final[normalize_col]
stock_y=dfA1final[label]
# Split the data into training, testing, and validation sets
stock_x_train,stock_x_test,stock_y_train,stock_y_test=train_test_split(stock_X,stock_y,test_size=0.15,random_state=42,shuffle=False)
stock_x_train,stock_x_val,stock_y_train,stock_y_val=train_test_split(stock_x_train,stock_y_train,test_size=0.15,random_state=42,shuffle=False)

# Initialize the Min-Max scaler for feature normalization
scaler=MinMaxScaler()
# Fit the Min-Max scaler to the training data
scaler.fit(stock_x_train)

# Transform the training, testing, and validation data using the Min-Max scaler
stock_x_train_scaled=scaler.transform(stock_x_train)
stock_x_test_scaled=scaler.transform(stock_x_test)
stock_x_val_scaled=scaler.transform(stock_x_val)
# Initialize the scaler for target variable normalization
scaler1=MinMaxScaler()
# Fit the scaler to the training target variable
scaler1.fit(stock_y_train)

# Transform the training, testing, and validation target variables using the Standard scaler
stock_y_train_scaled=scaler1.transform(stock_y_train)
stock_y_test_scaled=scaler1.transform(stock_y_test)
stock_y_val_scaled=scaler1.transform(stock_y_val)
#print(stock_y_train_scaled)
# Create DataFrames for the scaled feature sets
df_x_train=pd.DataFrame(stock_x_train_scaled,columns=[normalize_col])
df_x_test=pd.DataFrame(stock_x_test_scaled,columns=[normalize_col])
df_x_val=pd.DataFrame(stock_x_val_scaled,columns=[normalize_col])
# Create DataFrames for the scaled target variable sets
df_y_train=pd.DataFrame(stock_y_train_scaled,columns=[label])
df_y_test=pd.DataFrame(stock_y_test_scaled,columns=[label])
df_y_val=pd.DataFrame(stock_y_val_scaled,columns=[label])
```

```python
### LSTM model
import torch
import torch.nn as nn
# Define the LSTM model class
class Lstm(nn.Module):
    def __init__(self,input_size,hidden_size,nlayers,output_size):
        super(Lstm,self).__init__()
        # Initialize the number of LSTM layers and define the LSTM module
        self.nlayers=nlayers
        self.lstm=nn.LSTM(input_size,hidden_size,nlayers,batch_first=True)
        # Define the fully connected (linear) layer for output
        self.fc=nn.Linear(hidden_size,output_size)


    def forward(self,x):
        ##initialize hidden state 0
        # h0=torch.zeros(self.nlayers,x.size(0),self.hidden_size)
        # c0=torch.zeros(self.nlayers,x.size(0),self.hidden_size)
        # Pass the input sequence through the LSTM layers
        out, _=self.lstm(x)
        # Pass the LSTM output through the fully connected layer for final prediction
        out=self.fc(out)
        #print(out.size())
        return out
```

```python
# Import necessary libraries
from torch.utils.data import Dataset
# Define a custom dataset class
class CustomDataset(Dataset):
    def __init__(self, data, labels, length):
        # Initialize the dataset with data, labels, and sequence length
        self.data = data
        self.labels = labels
        self.length = length

    def __len__(self):
        # Define the length of the dataset
        return len(self.data) - self.length - 1

    def __getitem__(self, index):
        # Initialize empty lists for sequence data and labels
        seq_data = []
        out = []
        # Extract a sequence of data and corresponding labels
        seq_data.append(self.data.iloc[index:index+30, :].values.tolist())
        out.append(self.labels.iloc[index+1:index+31, :].values.tolist())
        # Flatten the output list and convert it to a NumPy array
        out = np.array(out).flatten()
        # Convert the sequence data and output to PyTorch tensors
        seq_data = torch.Tensor(seq_data)
        out = torch.Tensor(out)
        return seq_data, out
```

### Q (7)

##### 1. Buy Condition:If either the short-term moving average (short_avg) is less than the long-term moving average (long_avg)or the predicted price (pred) is greater than the current price (current).Additionally, ensure that the balance allows for buying at least one share (int(self.balance/current) > 0).If these conditions are met, the decision is set to "buy."
##### 2. Sell Condition:If either the short-term moving average (short_avg) is greater than the long-term moving average (long_avg)or the predicted price (pred) is less than the current price (current).Additionally, check if there are stocks held (self.stock_held > 0).If these conditions are met, the decision is set to "sell."
##### 3. Hold Condition:If none of the above conditions are met, the default decision is set to "hold."This means that if neither a strong buy nor a strong sell signal is present, the trader holds the current position.

```python
class trading_module:
    def __init__(self,balance,commission):
        # Initialize the trading module with balance and commission parameters
        self.balance = balance  
        self.stock_held = 0 
        self.commission=commission
    def buy(self,current):
        # Buy method for executing a buy trade
        if self.balance > 0:
            # Calculate the number of stocks you can buy
            num_to_buy = self.balance // current
            # Calculate the cost, including commissions
            cost = num_to_buy * current+self.commission*current
            # Update the balance and the number of stocks held
            self.balance -= cost
            self.stock_held += num_to_buy
            print(f"Bought {num_to_buy} stocks at ${current} each, balance ${self.balance}.")
        else:
            print("LB ")
    def sell(self,current):
        if self.stock_held > 0:
            # Calculate the revenue from selling stocks, minus commissions
            revenue = self.stock_held * current-self.commission*current
            self.balance += revenue
            print(f"Sold {self.stock_held} stocks at ${current} each , balance ${self.balance}.")
            self.stock_held = 0
        else:
            print("NS ")
    def hold(self):
        print("H ")
    def make_decision(self,short_avg,long_avg,pred,current):
        ## This is where I have implemeted muy Tradinig Startegy
        # 1. Buy Condition:
        # If either the short-term moving average (short_avg) is less than the long-term moving average (long_avg)
        # or the predicted price (pred) is greater than the current price (current).
        # Additionally, ensure that the balance allows for buying at least one share (int(self.balance/current) > 0).
        # If these conditions are met, the decision is set to "buy."
        if(((1.3*short_avg<long_avg) or (pred>1.1*current)) and int(self.balance/current)>0):
            decision="buy"
        # 2. Sell Condition:
        # If either the short-term moving average (short_avg) is greater than the long-term moving average (long_avg)
        # or the predicted price (pred) is less than the current price (current).
        # Additionally, check if there are stocks held (self.stock_held > 0).
        # If these conditions are met, the decision is set to "sell."
        elif(((short_avg>1.3*long_avg)or (pred<0.9*current)) and self.stock_held>0):
            decision="sell"
        # 3. Hold Condition:
        # If none of the above conditions are met, the default decision is set to "hold."
        # This means that if neither a strong buy nor a strong sell signal is present, the trader holds the current position.
        else:
            decision="hold"
        return decision
    def execute_trade(self,short_avg,long_avg,pred,current):
        decision = self.make_decision(short_avg,long_avg,pred,current)
        if decision == "buy":
            self.buy(current)
        elif decision == "sell":
            self.sell(current)
        elif decision == "hold":
            self.hold()
```
